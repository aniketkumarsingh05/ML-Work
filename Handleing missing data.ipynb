```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv", parse_dates=["day"])

df.set_index("day",inplace=True)
df
```

```python
#df.rename(columns = {'day' : 'date'})
```

```python
pd.isnull(df).sum()
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({'temperature': 0,
                  'windspeed':0,
                  'event': 'no event'})
new_df
```

```python
new_df=df.fillna(method="ffill")
new_df
```

```python
new_df=df.fillna(method="ffill", limit=2)
new_df
```

```python
new_df=df.fillna(method="bfill")
new_df
```

```python
new_df=df.fillna(method="bfill",axis="columns")
new_df
```

```python
new_df=df.interpolate(method='time')
new_df
```

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how="all")
new_df
#drop the row who has all non NA values
```

```python
new_df= df.dropna(thresh=1)
new_df
```

```python
new_df= df.dropna(thresh=2)
new_df
```

```python
new_df= df.dropna(thresh=3)
new_df
```

```python
dt=pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
#creating a date range of missing date
```
